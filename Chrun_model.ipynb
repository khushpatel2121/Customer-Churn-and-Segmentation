```python
import numpy as np
import pandas as pd
from  sklearn.decomposition import PCA 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers
```

```python
data = pd.read_csv("Encoded(3).csv")

features = data.drop(['Churn','CustomerID'],axis=1)
target = data['Churn']


#Standard Scaling 

scale =  StandardScaler()
scaled_features = scale.fit_transform(features)

#PCA for retaninig 95% of the variance of the data
pca = PCA(0.7)
pca_features = pca.fit_transform(scaled_features)

print(pca_features.shape)
print(target.shape)

X_train,X_test,Y_train,Y_test = train_test_split(pca_features,target,test_size=0.2,random_state=42)
```

```python


print(X_train.shape)
print(Y_train.shape)

#Neural Network 

model  = Sequential([
    Dense(67,activation='linear',input_shape=(pca_features.shape[1],),kernel_regularizer=regularizers.l2(0.01)),
    Dense(56,activation='linear',kernel_regularizer=regularizers.l2(0.01)),
    Dense(44,activation='linear',kernel_regularizer=regularizers.l2(0.01)),
    Dense(1,activation='sigmoid')
])

optimizer = Adam(learning_rate=0.0001)

model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])

model.fit(X_train,Y_train,epochs=100,batch_size=32,validation_data=(X_test,Y_test))

loss,accuracy = model.evaluate(X_test,Y_test)
print(f'Test Accuracy: {accuracy:.2f}')
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import accuracy_score

svm_model = SVC(kernel='rbf',C=1.0,gamma='scale',random_state=42)

svm_model.fit(X_train,Y_train)

y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(Y_test,y_pred)
print("Test Accuracy:",accuracy)
```

```python
#Decision Trees 
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Perform PCA
pca = PCA(n_components=0.95)  # Retain 95% variance
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Build Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)

# Train Decision Tree model
dt_model.fit(X_train_pca, y_train)

# Predict on the test set
y_pred = dt_model.predict(X_test_pca)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
```
